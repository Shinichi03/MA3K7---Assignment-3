```python
import random

def hat_numbers_simulation(n):
    numbers = list(range(1, n+1))
    while len(numbers) > 1:
        a, b = random.sample(numbers, 2) # Pick two random numbers
        new_number = abs(a - b)
        numbers.remove(a)
        numbers.remove(b)
        numbers.append(new_number) # Add the new number to the list
    return numbers[0] # Return the final number

# Run the simulation multiple times
n = 2025
final_numbers = [hat_numbers_simulation(n) for _ in range(500)] 
print(set(final_numbers))
```

```python
import matplotlib.pyplot as plt
from collections import Counter  

# Count occurrences of each final number
frequency_counts = Counter(final_numbers)

# Sort results for plotting
sorted_numbers = sorted(frequency_counts.keys())
sorted_freqs = [frequency_counts[num] for num in sorted_numbers]

# Plot the frequency distribution
plt.figure(figsize=(10, 5))
plt.bar(sorted_numbers, sorted_freqs, width=5, alpha=0.7, color='steelblue', edgecolor='black')
plt.xlabel("Final Remaining Number")
plt.ylabel("Frequency")
plt.title("Frequency Distribution of Final Remaining Number (n=2025)")
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```
